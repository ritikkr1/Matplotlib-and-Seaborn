```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D://ml//Car_sales.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Manufacturer'].fillna(value='audi',inplace=True)
df['x_4_year'].fillna(value=np.mean(df['x_4_year']),inplace=True)
df['Price in thousands'].fillna(value=np.mean(df['Price in thousands']),inplace=True)
df['Engine size'].fillna(value=np.mean(df['Engine size']),inplace=True)
df['Horsepower'].fillna(value=np.mean(df['Horsepower']),inplace=True)
df['Wheelbase'].fillna(value=np.mean(df['Wheelbase']),inplace=True)
df['Width'].fillna(value=np.mean(df['Width']),inplace=True)
df['Length'].fillna(value=np.mean(df['Length']),inplace=True)
df['Curb weight'].fillna(value=np.mean(df['Curb weight']),inplace=True)
df['Fuel capacity'].fillna(value=np.mean(df['Fuel capacity']),inplace=True)
df['Fuel efficiency'].fillna(value=np.mean(df['Fuel efficiency']),inplace=True)
df['XX'].fillna(value=np.mean(df['XX']),inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop('Unnamed: 15',axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.iloc[0:25,:]
```

# Matplotlib.pyplot

# line plot

```python
plt.figure(figsize=(21,9))
plt.plot(df['Model'],df['Sales in thousands'],'co-',markersize=10)
plt.title("sales in thousand as per model",fontsize=15)
plt.xlabel("Model",fontsize=15)
plt.ylabel("sales in thousand",fontsize=15)
plt.legend(['sales'])
plt.show()
```

# histrogram

```python
plt.figure(figsize=(21,9))
plt.hist(df['Model'])
plt.title("Model-Type",fontsize=15)
plt.xlabel("Model",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(21,9))
plt.hist(df['Horsepower'],rwidth=0.8)
plt.title("Horsepower",fontsize=15)
plt.xlabel("Horsepower",fontsize=15)
plt.ylabel("frequency",fontsize=15)
plt.show()
```

# Barchart

```python
plt.figure(figsize=(21,9))
plt.bar(df['Model'],df['Fuel efficiency'])
plt.xlabel('Model',fontsize=15)
plt.ylabel('Fuel efficiency',fontsize=15)
```

# scatter plot

```python
plt.figure(figsize=(21,9))
plt.scatter(df['Manufacturer'],df['Model'])
plt.xlabel('Manufacturer',fontsize=15)
plt.ylabel('Model',fontsize=15)
```

```python
plt.figure(figsize=(21,9))
plt.scatter(df['Model'],df['Price in thousands'])
plt.xlabel('Model',fontsize=15)
plt.ylabel('Price in thousands',fontsize=15)
```

# piechart

```python
plt.pie(df['Engine size'],autopct='%f%%',radius=2)
plt.show()
```
